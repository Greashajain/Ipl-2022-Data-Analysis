#**Loading the Libraries and Dataset**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/IPL.csv')
```

```python
df.head()
```

##**Basis Information**

```python
print(f"your rows are {df.shape[0]} and your columns are { df.shape[1]}")
```

```python
#now see how many columns have null values
df.isnull().sum()
```

```python
#which team wins the most matches
match_wins=df['match_winner'].value_counts()
```

```python
match_wins
```

```python
sns.barplot(y=match_wins.index,x= match_wins.values,palette='viridis')
plt.title('Most match wins by teams')
```

```python
#Toss decision Trend
sns.countplot(x=df['toss_decision'],palette='viridis')
plt.title('Toss Decision Trend')
```

```python
#Toss winner v/s Match winner
count_wins=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count_wins*100)/df.shape[0]
percentage.round(2)
```

```python
# How do teams win (Run v/s Wickets)
#df['won_by'].value_counts()
sns.countplot(x=df['won_by'],palette='viridis')
```

##**Key Player Performance**

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(y=count.index,x=count.values,palette='viridis')
plt.title('Top 10 Player of the Match Awards')
```

```python
# 2 Top Scorers
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh')
plt.title('Top Scorers')
```

```python
# 10 best Bowling Figures
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind = 'barh')
```

##**Venue Analysis**

```python
venue_count = df['venue'].value_counts()
venue_count
```

```python
sns.barplot(y = venue_count.index,x = venue_count.values,palette='rainbow')
```
